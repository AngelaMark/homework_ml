```python
data = data.dropna(axis=1)
```

```python
data.count()
```

```python
del data["ID"]
del data["Source"]
del data["Location Source"]
del data["Magnitude Source"]
```

```python
data.sample(3)
```

```python
data["Datetime"] = pd.to_datetime(data["Date"]).map(str) + pd.to_datetime(data["Time"]).map(str)
data["Datetime"] =  pd.to_datetime(data["Datetime"])
data["Datetime"] = pd.to_numeric(data["Datetime"])
del data["Time"]
del data["Date"]
y = data["Magnitude"]
del data["Magnitude"]
```

```python
from sklearn.preprocessing import LabelEncoder
lenc = LabelEncoder()
data["Type"] = lenc.fit_transform(data["Type"])
data["Status"] = lenc.fit_transform(data["Status"])
```

```python
data.sample(100)
```

```python
from sklearn.model_selection import train_test_split
X = data.values
```

```python
from sklearn import tree
from sklearn.cross_validation import cross_val_score
from sklearn.cross_validation import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.4,random_state=0)

random_grid={
    'n_estimators': [60, 65, 70],

    'max_depth' : [5,8,10],
}
from sklearn.ensemble import RandomForestRegressor


tree_r = GridSearchCV(RandomForestRegressor(), random_grid, cv=10,refit=True,
                    verbose=1,n_jobs=4,error_score='mse',return_train_score='mse')
tree_r.fit(X_train, y_train)
```

```python
tree_r.best_params_
```

```python
y_pred = tree_r.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error

error = mean_squared_error(y_test, y_pred)
```
